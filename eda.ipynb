```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = sns.load_dataset('titanic')

# Check data types
print(df.dtypes)

# Select only numeric columns
df_numeric = df.select_dtypes(include=[float, int])

# Generate the pairplot
sns.pairplot(df_numeric, hue='survived')

# Display the plot
plt.show()
```

```python
#stacked barchart/plot
df.groupby(['pclass','who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarm
sns.swarmplot(data=df, x='survived',y='age',hue='who')
```

```python
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
# relation of two numeric variable
sns.lmplot(data=df, y='fare',x='age')
```

```python
sns.lmplot(data=df, x='age',y='fare', hue='who')
plt.ylim(0,300)
```

```python
# how to add a column and make a age group
sns.histplot(data=df, x='age')
```

```python
# grouping
df.columns
```

```python
df['age_groups']= pd.cut(df['age'], bins= [0,10,18,30,50,80],
    labels=['kids','child','youth','middle aged','old'])
```
